```python
import plotly.graph_objects as go
import numpy as np

# Define nodes and connections
nodes = [
    "Transit Signal Priority (TSP)", "AI-RAN / NVIDIA Aerial", "Smart Cities",
    "Urban Mobility", "Edge AI Infrastructure", "Intersection Delay Reduction",
    "Bus Priority at Signals", "Cellular Connectivity", "Edge AI at Intersections",
    "5G Connectivity", "Edge AI Inference", "Cloud Orchestration", "Low-latency Optimization",
    "Connected Infrastructure", "Autonomous Vehicles", "Predictive Traffic Models",
    "Public Transit OTP", "Bike & Pedestrian Prioritization", "Multimodal Integration",
    "Signal Cabinets with Compute Power", "Real-time Local Decision-Making", "AI Chips (e.g., NVIDIA Jetson)"
]

edges = [
    ("Transit Signal Priority (TSP)", "Intersection Delay Reduction"),
    ("Transit Signal Priority (TSP)", "Bus Priority at Signals"),
    ("Transit Signal Priority (TSP)", "Cellular Connectivity"),
    ("Transit Signal Priority (TSP)", "Edge AI at Intersections"),
    ("AI-RAN / NVIDIA Aerial", "5G Connectivity"),
    ("AI-RAN / NVIDIA Aerial", "Edge AI Inference"),
    ("AI-RAN / NVIDIA Aerial", "Cloud Orchestration"),
    ("AI-RAN / NVIDIA Aerial", "Low-latency Optimization"),
    ("Smart Cities", "Connected Infrastructure"),
    ("Smart Cities", "Autonomous Vehicles"),
    ("Smart Cities", "Predictive Traffic Models"),
    ("Urban Mobility", "Public Transit OTP"),
    ("Urban Mobility", "Bike & Pedestrian Prioritization"),
    ("Urban Mobility", "Multimodal Integration"),
    ("Edge AI Infrastructure", "Signal Cabinets with Compute Power"),
    ("Edge AI Infrastructure", "Real-time Local Decision-Making"),
    ("Edge AI Infrastructure", "AI Chips (e.g., NVIDIA Jetson)"),
    ("Cellular Connectivity", "5G Connectivity"),
    ("Edge AI at Intersections", "Edge AI Inference"),
    ("Cloud Orchestration", "Connected Infrastructure"),
    ("Predictive Traffic Models", "Public Transit OTP"),
    ("AI Chips (e.g., NVIDIA Jetson)", "Edge AI Inference"),
    ("Multimodal Integration", "Bike & Pedestrian Prioritization")
]

# Circular layout
num_nodes = len(nodes)
angle_step = 360 / num_nodes
positions = [(i, angle_step * i) for i in range(num_nodes)]
x = [1.5 * np.cos(np.radians(angle)) for i, angle in positions]
y = [1.5 * np.sin(np.radians(angle)) for i, angle in positions]

# Node trace
node_trace = go.Scatter(
    x=x, y=y,
    mode='markers+text',
    text=nodes,
    textposition="top center",
    marker=dict(size=20, color='lightblue', line=dict(width=2, color='darkblue')),
    hoverinfo='text'
)

# Edge traces
edge_traces = []
for edge in edges:
    start = nodes.index(edge[0])
    end = nodes.index(edge[1])
    edge_trace = go.Scatter(
        x=[x[start], x[end], None],
        y=[y[start], y[end], None],
        mode='lines',
        line=dict(width=2, color='gray'),
        hoverinfo='none'
    )
    edge_traces.append(edge_trace)

# Plot it
fig = go.Figure(data=edge_traces + [node_trace],
                layout=go.Layout(
                    title='Interactive Mind Map: TSP, AI-RAN, Smart Mobility',
                    titlefont_size=16,
                    showlegend=False,
                    hovermode='closest',
                    margin=dict(b=20,l=5,r=5,t=40),
                    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False)
                ))
fig.show()
```
